```python
import numpy as np
import pandas as pd

df = pd.DataFrame({'x':[7,2,0,3,4,2,5,0,3,4]})
data = df['x']
new_list =[]
x=0
for value in data:
    if value !=0:
        x = x+1
    else:
        x=0
        
    new_list.append(x) 
print(new_list)
df['value']=new_list
df
```

```python
dti = pd.bdate_range('2015-01-01','2015-12-31')
dti1 = pd.DataFrame({'date':dti})
rand=np.random.randint(1,500,len(dti1))

rand_df = pd.DataFrame({'Values':rand},index=dti)

rand_df
```

```python
from datetime import date
from datetime import time
days =['Monday','Tuesday','Wednesday','Thursday','Friday']
day_list=[]
total =0
for x in dti:
    s=date.weekday(x)
    val = days[s]
    day_list.append(val)
rand_df['Day']=day_list
rand_df
rand_df_wednesday = rand_df[rand_df['Day']== 'Wednesday']
for x in rand_df_wednesday['Values']:
    total = total+x
Average = total/len(rand_df_wednesday)
Average
print(rand_df_wednesday)
print("Total = %f"%total)
print("Average Value for Wednesday: %f"%Average)
```
