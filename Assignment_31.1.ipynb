```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
iris_pd=pd.DataFrame(iris.data)
iris_pd.head()
```

```python
x=iris.data[:,:2]
y=iris.target
```

```python
x_min, x_max = x[:, 0].min() - .5, x[:, 0].max() + .5
y_min, y_max = x[:, 1].min() - .5, x[:, 1].max() + .5
plt.figure(2, figsize=(8, 6))
plt.clf()
plt.scatter(x[:, 0], x[:, 1], c=y, cmap=plt.cm.Set1,edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```

```python
ax = Axes3D(plt.figure(1, figsize=(8, 6)), elev=-150, azim=110)
x_red = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(x_red[:, 0],x_red[:, 1],x_red[:, 2], c=y,cmap=plt.cm.Set1, edgecolor='b', s=40)
ax.set_title("PCA 3D ANALYSIS")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])
plt.show()
```
