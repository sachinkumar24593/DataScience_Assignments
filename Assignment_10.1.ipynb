```python
len=dataset.Name.count()
print(len)
if(len%2==0):
    median = len/2
    print(dataset.iloc[median-1])
    median1 = (len/2)+1
    print(dataset.iloc[median1-1])
else:
    median = int((len+1)/2)
    print(dataset.iloc[median-1])
```

```python
import pandas as pd
ser = pd.Series()
df = pd.DataFrame(dataset)
df_count =df['Name'].value_counts()
list1 = list(df_count)
for x in range(0,5):
    maxval = df_count[df_count == max(list1)]
    ser = ser.append(maxval)
    list1.remove(max(list1))
df_max=pd.DataFrame({'Maxvalue':ser})
list_name = list(df_max.index.values)
df_max['Name']=list_name
df_max.index=[1,2,3,4,5]
df_max

```
