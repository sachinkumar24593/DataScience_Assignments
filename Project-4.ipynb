# Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib as plt
import pylab as py
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

# DataSet

```python
train_data=pd.read_csv('f:/Downloads/data_stocks.csv')
```

```python
train_data.head()
```

# Implementation of Principal Component Analysis

```python
pca=PCA(n_components=3)
pca.fit(train_data)
x_red=pca.transform(train_data)
x_red.shape
```

```python
x_red
```

```python
py.scatter(x_red[:,0],x_red[:,1],cmap='RdYlBu')
py.show()
```

# Implementation of Kmeans for Actual Dataset

```python
k_means=KMeans(n_clusters=3,random_state=0)
k_means.fit(train_data)
y_pred=k_means.predict(train_data)
```

# Applying Cluster to the reduced Data (PCA Data)

```python
py.scatter(x_red[:,0],x_red[:,1],c=y_pred,cmap='RdYlBu')
```
