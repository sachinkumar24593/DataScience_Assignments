```python
users_df=pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/users.csv')
users_df.head()
```

```python
session_df=pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/sessions.csv')
session_df.head()
```

```python
products_df=pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/products.csv')
products_df.head()
```

```python
transaction_df=pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv')
transaction_df
```

```python
new_df=transaction_df.merge(users_df, left_on='UserID', right_on='UserID', how='left')
new_df
```

```python
nan_df=new_df[new_df['User'].isnull()]
nan_df
```

```python
transaction_df.merge(users_df, left_on='UserID', right_on='UserID', how='inner')
```

```python
transaction_df.merge(users_df, left_on='UserID', right_on='UserID', how='outer')
```

```python
users_df.merge(session_df, left_on='Registered', right_on='SessionDate', how='inner')
```

```python
index = pd.MultiIndex.from_product([users_df.UserID,products_df.ProductID], names = ["UserID", "ProductID"])
cross_df = pd.DataFrame(index = index).reset_index()
cross_df
```

```python
qlist=[]
for x in range(1,6):
    for y in range(1,6):
        if(transaction_df[(transaction_df['UserID']==x)&(transaction_df['ProductID']==y)].empty):
            qlist.append(0)
        else:
            qlist.append(transaction_df[(transaction_df['UserID']==x)&(transaction_df['ProductID']==y)].Quantity.values.sum())
cross_df["Quantity"]=qlist 
cross_df
```

```python
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
nw_df=pysqldf('''SELECT * FROM ( SELECT * FROM transaction_df ORDER BY TransactionDate desc) as temp GROUP BY temp.UserID;''')
nw_df
nw1_df=nw_df.dropna()
nw2_df=nw1_df[nw1_df['UserID']<6]
nw2_df

```

```python
users_df.merge(nw2_df, left_on='UserID',right_on='UserID',how='left')
```
