```python
import numpy as np
import pandas as pd
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm',
'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
df
```

```python
listn=(df[df.FlightNumber.isnull()].index)
for n in listn:
    val=df.iloc[n-1].FlightNumber+10
    df.set_value(n,'FlightNumber',val)
df
```

```python
fromlist=[]
tolist=[]
ftlist=df['From_To']
for st in ftlist:
    t= st.split('_')
    fromlist.append(t[0].title())
    tolist.append(t[1].title())
df['From']=fromlist
df['To']=tolist
df
```

```python
df.drop(columns=['From_To'])
```

```python
delay1=[]
delay2=[]
delay3=[]
templist=df.RecentDelays
for n in templist:
    if(len(n)!=0):
        for x in range(0,3):
            if(x<len(n)):
                if(x==0):
                    delay1.append(n[x])
                if(x==1):
                    delay2.append(n[x])
                if(x==2):
                    delay3.append(n[x])
            else:
                if(x==1):
                    delay2.append(np.nan)
                if(x==2):
                    delay3.append(np.nan)
    else:
        delay1.append(np.nan)
        delay2.append(np.nan)
        delay3.append(np.nan)
        
df['delay_1']=delay1
df['delay_2']=delay2
df['delay3']=delay3
df
```
